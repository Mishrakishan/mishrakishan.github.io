# Superposition State
$|\psi\rangle = \frac{1}{\sqrt{2}}\left(|0\rangle+1\rangle\right)$

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit.visualization import plot_bloch_multivector, plot_state_city, plot_state_qsphere
from qiskit.quantum_info import Statevector, DensityMatrix
import matplotlib.pyplot as plt

# --- Superposition State ---
# Create a quantum circuit for the superposition state
qreg_q = QuantumRegister(1, 'q')
creg_c = ClassicalRegister(1, 'c')
circuit_superposition = QuantumCircuit(qreg_q, creg_c)

# Apply Hadamard gate to create superposition state |+> = 1/sqrt(2) (|0> + |1>)
circuit_superposition.h(qreg_q[0])

# Calculate the statevector
psi_superposition = Statevector(circuit_superposition)
```

```python
circuit_superposition.draw(output="mpl")
```

```python
plot_bloch_multivector(psi_superposition)
```

```python

plot_state_city(psi_superposition)
```

```python

plot_state_qsphere(psi_superposition)
```

```python
# Plot Probability Amplitudes (magnitude squared) for superposition state
probabilities = np.abs(psi_superposition.data)**2
plt.subplot(2, 3, 4)
plt.bar(range(len(probabilities)), probabilities, color='b', alpha=0.7)
plt.title('Probability Amplitudes (Superposition)')
plt.xlabel('State |0>, |1>')
plt.ylabel('Probability')
```

```python
plt.subplot(2,3,5)
plt.bar(range(len(psi_superposition.data)), np.real(psi_superposition.data), color='r', alpha=0.7, label='Real part')
plt.bar(range(len(psi_superposition.data)), np.imag(psi_superposition.data), color='b', alpha=0.7, label='Imaginary part')
plt.title('State Vector (Superposition)')
plt.xlabel('State |0>, |1>')
plt.ylabel('Amplitude')
plt.legend()
```

#Entangled State (Bell State)
$|\phi^{+}\rangle = \frac{1}{\sqrt{2}}\left(|00\rangle+11\rangle\right)$

```python
# --- Entangled State (Bell State) ---
# Create a quantum circuit for the entangled state (Bell state)
qreg_q1 = QuantumRegister(1, 'q1')
qreg_q2 = QuantumRegister(1, 'q2')
creg_c1 = ClassicalRegister(1, 'c1')
creg_c2 = ClassicalRegister(1, 'c2')
circuit_entangled = QuantumCircuit(qreg_q1, qreg_q2, creg_c1, creg_c2)

# Apply Hadamard gate on the first qubit
circuit_entangled.h(qreg_q1[0])

# Apply CNOT gate to entangle the qubits
circuit_entangled.cx(qreg_q1[0], qreg_q2[0])
# Calculate the statevector for the entangled state
psi_entangled = Statevector(circuit_entangled)
```

```python
circuit_entangled.draw(output="mpl")
```

```python
# Plot Bloch sphere for entangled state
plot_bloch_multivector(psi_entangled)
```

```python
# Plot State City (amplitude and phase) for entangled state
plot_state_city(psi_entangled)
```

```python

# Plot State Q-sphere for entangled state
plot_state_qsphere(psi_entangled)
```

```python
# Plot Probability Amplitudes (magnitude squared) for entangled state
probabilities_entangled = np.abs(psi_entangled.data)**2
plt.subplot(2, 3, 4)
plt.bar(range(len(probabilities_entangled)), probabilities_entangled, color='g', alpha=0.7)
plt.title('Probability Amplitudes (Entangled State)')
plt.xlabel('State |00>, |01>, |10>, |11>')
plt.ylabel('Probability')
```

```python
# Plot State Vector (real and imaginary parts) for entangled state
plt.subplot(2, 3, 5)
plt.bar(range(len(psi_entangled.data)), np.real(psi_entangled.data), color='r', alpha=0.7, label='Real part')
plt.bar(range(len(psi_entangled.data)), np.imag(psi_entangled.data), color='b', alpha=0.7, label='Imaginary part')
plt.title('State Vector (Entangled State)')
plt.xlabel('State |00>, |01>, |10>, |11>')
plt.ylabel('Amplitude')
plt.legend()
```

## Quantum Gates

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit.visualization import plot_bloch_multivector, plot_state_city, plot_state_qsphere
from qiskit.quantum_info import Statevector
import matplotlib.pyplot as plt
import numpy as np

# Create a quantum circuit for a single qubit state |0> initially
qreg_q = QuantumRegister(1, 'q')
creg_c = ClassicalRegister(1, 'c')
circuit = QuantumCircuit(qreg_q, creg_c)

# --- Quantum Gate Applications ---
# Apply different gates on the qubit and visualize the results

# Identity Gate (No operation)
circuit.id(qreg_q[0])

# Apply Hadamard gate (creates superposition state)
circuit.h(qreg_q[0])

# Apply Pauli-X gate (bit-flip)
circuit.x(qreg_q[0])

# Apply Pauli-Y gate
circuit.y(qreg_q[0])

# Apply Pauli-Z gate
circuit.z(qreg_q[0])

# Apply S gate (Phase gate)
circuit.s(qreg_q[0])

# Apply T gate (π/8 gate)
circuit.t(qreg_q[0])

# Apply CNOT gate (entangling two qubits)
qreg_q2 = QuantumRegister(1, 'q2')
circuit.add_register(qreg_q2)
circuit.cx(qreg_q[0], qreg_q2[0])

# Create statevector from circuit
psi = Statevector(circuit)
```

```python
circuit.draw(output = "mpl")
```

```python
# Plot Bloch Sphere
plot_bloch_multivector(psi)
```

```python
# Plot State City (amplitude and phase)
plot_state_city(psi)
```

```python

plot_state_qsphere(psi)

```

```python
# Plot Probability Amplitudes
plt.subplot(2,3,2)
probabilities = np.abs(psi.data)**2
plt.bar(range(len(probabilities)), probabilities, color='g', alpha=0.7)
plt.title('Probability Amplitudes')
plt.xlabel('State |0>, |1>, |00>, |01>, |10>, |11>')
plt.ylabel('Probability')
```
